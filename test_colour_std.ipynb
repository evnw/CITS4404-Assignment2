## Testing How To Get Colour Under Worm

```python
# imports
from camo_worms_utils import *
```

```python
img = prep_image(IMAGE_DIR, IMAGE_NAME, MASK)
```

```python
# what is the mean colour
print(np.mean(img)/255)
```

Value of ~0.2 for mean colour => stdev is on bounds [0, ~0.16]. See [here](https://math.stackexchange.com/questions/265360/restrictions-on-a-bounded-variables-mean-and-standard-deviation#:~:text=Given%20a%20random%20variable%20whose,not%20every%20combination%20is%20possible.)

So lets use scaling factor 1/0.16=6.25 to scale the bounds to [0, 1]

```python
worm1 = Camo_Worm(200, 100, 60, 0, 90, 30, 20, 0.6)
worm2 = Camo_Worm(350, 170, 10, 0, 0, 0, 10, 0.6)
drawing = Drawing(img)
drawing.add_worms([worm1, worm2])
drawing.show()
```

```python
colours = worm1.get_colour_under(img)
std = np.std(colours)*6.25

print(f"Colour Standard Deviation: {std}")

# draw reference worm over old one with avg colour
drawing = Drawing(img)
drawing.add_worms([worm1, worm2])
drawing.add_dots(worm1.intermediate_points(3), 2, color="yellow")
drawing.show()
```

```python
colours = worm2.get_colour_under(img)
std = np.std(colours)*6.25

print(f"Colour Standard Deviation: {std}")

# draw reference worm over old one with avg colour
drawing = Drawing(img)
drawing.add_worms([worm1, worm2])
drawing.add_dots(worm2.intermediate_points(3), 2, color="yellow")
drawing.show()
```
